spaCy organizes its NLP tasks into a processing pipeline. The pipeline typically involves tokenization, part of speech tagging, named entity recognition and other linguistic analyses. When you load a spaCy model using spaCy.load, you get an NLP object that contains the language model and processing pipeline. When you call NLP on a text, spaCy first tokenizes the text to produce a doc object. The doc is then processed and several different steps. This is also referred to as the processing pipeline.

```python
import spacy


# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Process a text using the nlp object
doc = nlp("This is an example sentence.")

# Access linguistic annotations
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

Each token in the sentence is printed along with its associated dependency label. The output shows the grammatical relationships between the tokens in the sentence.

Dependency parsing analyzes how words are grammatically connected in a sentence.

| Dependency | Meaning                                                                          |
| ---------- | -------------------------------------------------------------------------------- |
| `ROOT`     | The main verb or core of the sentence                                            |
| `nsubj`    | **Nominal subject** – the person or thing doing the action                       |
| `attr`     | **Attribute** – something being described or stated about the subject            |
| `det`      | **Determiner** – words like "the", "a", "an"                                     |
| `compound` | **Compound modifier** – a noun modifying another noun (e.g., "example sentence") |
| `dobj`     | **Direct object** – the thing being acted upon                                   |
| `amod`     | **Adjective modifier** – adjective describing a noun                             |
| `prep`     | **Prepositional modifier** – introduces a phrase like "in the box"               |
| `pobj`     | **Object of a preposition** – the noun inside the prepositional phrase           |
| `punct`    | Punctuation – connects to the main clause for sentence completeness              |

**NER(Named Entity Recognition)**

 In named entity recognition, you extract out the name of the person, store, or location.

```python
import spacy

# Load the English NLP model
nlp3 = spacy.load("en_core_web_sm")

# Define a sample text
text = ("OnlineStore,Inc. is planning to open a new store in San Francisco. "
        "Core Greene, the CEO, will be attending the launch event.")

# Process the text using spacy
doc = nlp3(text)

# Extract named entities
entities = [(ent.text, ent.label_) for ent in doc.ents]

# Print named entities
print("Named entities:")
for entity, label in entities:
    print(f"{entity} - {label}")
```

An English NLP model is a pre-trained machine learning model designed to process and analyze English text. These models are trained on large datasets to perform various natural language processing (NLP) tasks, such as:

Common Tasks Supported by English NLP Models
Tokenization: Splitting text into words, punctuation, etc.

Example: "Hello, world!" → ["Hello", ",", "world", "!"]

Part-of-Speech (POS) Tagging: Identifying grammatical roles (noun, verb, adjective, etc.).

Example: "run" → VERB (in "I run") or NOUN (in "a quick run").

Named Entity Recognition (NER): Detecting names of people, organizations, locations, etc.

Example: "Apple is based in Cupertino." → Apple (ORG), Cupertino (GPE).

Dependency Parsing: Analyzing grammatical sentence structure.

Example: Identifying subjects, objects, and relationships.

Lemmatization: Reducing words to their base/dictionary form.

Example: "running" → "run", "better" → "good".

Sentence Segmentation: Splitting text into sentences.

Example: "Hello! How are you?" → 2 sentences.

**POS Tagging**

 Part of speech tagging involve understanding the structure and meaning of natural language text. For example, identifying parts of speech helps identify potential named entities and their roles in a sentence.

```python
import spacy

# Load the English NLP model
nlp = spacy.load("en_core_web_sm")

# Define the text
text = ("Your products are excellent. "
        "However, delivery to my location in New York was a problem.")

# Process the text using spaCy
doc = nlp(text)

# Extract POS tags and print them
print("Part-of-Speech Tags:")
for token in doc:
    print(f"{token.text} = {token.pos_}")

# Extract named entities and print them
print("\nNamed Entities:")
for ent in doc.ents:
    print(f"{ent.text} = {ent.label_}")
```

```python
# Tokenization
print("Tokenization:")
for token in doc:
    # Iterate through each token in the processed text and print the token text
    print(token.text)

print("\n")  # Add a newline for separation

# Filter out stop words
print("Filtered Tokens (without stop words):")
filtered_tokens = [token.text for token in doc if not token.is_stop]
# Create a list of tokens excluding stop words using list comprehension
print(filtered_tokens)

print("\n")  # Add a newline for separation

# Part-of-Speech Tagging (POS)
print("Part-of-Speech Tagging (POS):")
for token in doc:
    # Iterate through each token and print the token text and its POS tag
    print(token.text, token.pos_)

print("\n")  # Add a newline for separation

# Named Entity Recognition (NER)
print("Named Entity Recognition (NER):")
for ent in doc.ents:
    # Iterate through each named entity in the processed text and print its text and label
    print(ent.text, ent.label_)

print("\n")  # Add a newline for separation

# Lemmatization
print("Lemmatization:")
lemmatized_tokens = [token.lemma_ for token in doc if not token.is_punct]
# Create a list of lemmatized tokens excluding punctuation using list comprehension
print(lemmatized_tokens)
```
